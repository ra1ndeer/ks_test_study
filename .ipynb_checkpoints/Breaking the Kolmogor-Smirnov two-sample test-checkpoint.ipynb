```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from scipy import stats
sns.set_style("whitegrid")
```

## Mean difference detection

```python
sample_sizes = [10, 30, 50, 100, 150, 200, 500, 1000]

min_mean = -3
max_mean = 3
means = np.linspace(min_mean, max_mean, granularity)

granularity = 50
num_iterations = 100

df_list = list()

for sample_size in sample_sizes:
    df = pd.DataFrame(data=means, columns=["mean_data"])
    
    for i in range(num_iterations):
        ks_results = list()
        
        for mean in means:
            s0 = np.random.normal(loc=0, scale=1, size=sample_size)
            s_test = np.random.normal(loc=mean, scale=1, size=sample_size)
            ks_results.append(stats.ks_2samp(s0, s_test)[1])
            
        df["iter_"+str(i)] = ks_results
        
    df["mean_res"] = df[df.columns[1:]].mean(axis=1)
    df["std_res"] = df[df.columns[1:]].std(axis=1)
    df_list.append(df)
```

```python
for data_df in df_list:
    fig, ax = plt.subplots(figsize=(25, 5))

    plt.errorbar(data_df["mean_data"], 
                 data_df["mean_res"], 
                 yerr=data_df["std_res"], 
                 elinewidth=0.7,
                 linestyle="--", 
                 marker=".", ecolor="red", alpha=0.5, color="black")
    
    ax.fill_between([min_mean, max_mean], [0.01, 0.01], [0, 0], alpha=0.2, label="0.01")
    ax.fill_between([min_mean, max_mean], [0.05, 0.05], [0.01, 0.01], alpha=0.2, label="0.05")
    ax.fill_between([min_mean, max_mean], [0.1, 0.1], [0.05, 0.05], alpha=0.2, label="0.1")
    
    ax.set_xlim((min_mean, max_mean))
    ax.set_ylim((0, 1))
    
    ax.set_xlabel("Normal distribution mean")
    ax.set_ylabel("p-value")
    ax.legend(title="Confidence levels")
    
plt.show()
```

## Standard deviation difference detection

```python
min_scale = 0.5
max_scale = 1.5

sample_size = 1000
granularity = 300
num_iterations = 1000

scales = np.linspace(min_scale, max_scale, granularity)
df_std_simulation = pd.DataFrame(data=scales, columns=["std_data"])
```

```python
# simulation loop
for i in range(num_iterations):
    ks_results = list()
    
    # calculating the KS test p-values for each standard deviation value
    for scale in scales:
        s0 = np.random.normal(loc=0, scale=1, size=sample_size)
        s_test = np.random.normal(loc=0, scale=scale, size=sample_size)
        ks_results.append(stats.ks_2samp(s0, s_test)[1])
    
    # appending the results to the final dataframe
    df_std_simulation["iter_"+str(i)] = ks_results
```

```python
# calculate resulting mean and standard deviation values
df_std_simulation["mean_res"] = df_std_simulation[df_std_simulation.columns[1:]].mean(axis=1)
df_std_simulation["std_res"] = df_std_simulation[df_std_simulation.columns[1:]].std(axis=1)
```

```python
# plotting the results
fig, ax = plt.subplots(figsize=(16, 10))

plt.errorbar(df_std_simulation["std_data"], 
             df_std_simulation["mean_res"], 
             yerr=df_std_simulation["std_res"], 
             elinewidth=0.7,
             linestyle="--", 
             marker=".", ecolor="red", alpha=0.5, color="black")

ax.fill_between([min_scale, max_scale], [0.01, 0.01], [0, 0], alpha=0.2, label="0.01")
ax.fill_between([min_scale, max_scale], [0.05, 0.05], [0.01, 0.01], alpha=0.2, label="0.05")
ax.fill_between([min_scale, max_scale], [0.1, 0.1], [0.05, 0.05], alpha=0.2, label="0.1")

ax.set_xlim((min_scale, max_scale))
ax.set_ylim((0, 1))

ax.set_xlabel("Normal distribution standard deviation")
ax.set_ylabel("p-value")
ax.legend(title="Confidence levels")

plt.show()
```

## Sample size sensitivity

```python
min_size = 10
max_size = 5000
test_mean = 0.1

num_iterations = 1000

sizes = np.arange(min_size, max_size, step=5)
df_size_simulation = pd.DataFrame(data=sizes, columns=["size"])
```

```python
# simulation loop
for i in range(num_iterations):
    ks_results = list()
    
    # calculating the KS test p-values for each standard deviation value
    for s in sizes:
        s0 = np.random.normal(loc=0, scale=1, size=s)
        s_test = np.random.normal(loc=test_mean, scale=1, size=s)
        ks_results.append(stats.ks_2samp(s0, s_test)[1])
    
    # appending the results to the final dataframe
    df_size_simulation["iter_"+str(i)] = ks_results
```

```python
# calculate resulting mean and standard deviation values
df_size_simulation["mean_res"] = df_size_simulation[df_size_simulation.columns[1:]].mean(axis=1)
df_size_simulation["std_res"] = df_size_simulation[df_size_simulation.columns[1:]].std(axis=1)
```

```python
# plotting the results
fig, ax = plt.subplots(figsize=(16, 10))

plt.errorbar(df_size_simulation["size"], 
             df_size_simulation["mean_res"], 
             yerr=df_size_simulation["std_res"], 
             elinewidth=0.7,
             linestyle="--", 
             marker=".", ecolor="red", alpha=0.3, color="black")

ax.fill_between([min_size, max_size], [0.01, 0.01], [0, 0], alpha=0.2, label="0.01")
ax.fill_between([min_size, max_size], [0.05, 0.05], [0.01, 0.01], alpha=0.2, label="0.05")
ax.fill_between([min_size, max_size], [0.1, 0.1], [0.05, 0.05], alpha=0.2, label="0.1")

ax.set_xlim((min_size, max_size))
ax.set_ylim((0, 1))

ax.set_xlabel("Sample size")
ax.set_ylabel("p-value")
ax.legend(title="Confidence levels")

plt.show()
```
