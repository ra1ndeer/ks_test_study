```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from scipy import stats
sns.set_style("whitegrid")
```

## Mean difference detection

### Simulation

```python
sample_sizes = [15, 30, 50, 100, 150, 200, 500, 1000]

min_mean = -3
max_mean = 3

granularity = 300
num_iterations = 1000

means = np.linspace(min_mean, max_mean, granularity)

df_list = list()

for sample_size in sample_sizes:
    df = pd.DataFrame(data=means, columns=["mean_data"])
    
    for i in range(num_iterations):
        ks_results = list()
        
        for mean in means:
            s0 = np.random.normal(loc=0, scale=1, size=sample_size)
            s_test = np.random.normal(loc=mean, scale=1, size=sample_size)
            ks_results.append(stats.ks_2samp(s0, s_test)[1])
            
        df["iter_"+str(i)] = ks_results
        
    df["mean_res"] = df[df.columns[1:]].mean(axis=1)
    df["std_res"] = df[df.columns[1:]].std(axis=1)
    df_list.append(df)
```

### Plotting the results

```python
for df in df_list:
    fig, ax = plt.subplots(figsize=(25, 5))

    plt.errorbar(df["mean_data"], 
                 df["mean_res"], 
                 yerr=df["std_res"], 
                 elinewidth=0.7,
                 linestyle="--", 
                 marker=".", ecolor="red", alpha=0.5, color="black")
    
    ax.fill_between([min_mean, max_mean], [0.01, 0.01], [0, 0], alpha=0.2, label="0.01")
    ax.fill_between([min_mean, max_mean], [0.05, 0.05], [0.01, 0.01], alpha=0.2, label="0.05")
    ax.fill_between([min_mean, max_mean], [0.1, 0.1], [0.05, 0.05], alpha=0.2, label="0.1")
    
    ax.set_xlim((min_mean, max_mean))
    ax.set_ylim((0, 1))
    
    ax.set_xlabel("Normal distribution mean")
    ax.set_ylabel("p-value")
    ax.legend(title="Confidence levels")
    
plt.show()
```

### Saving the data to files

```python
for df, sample_size in zip(df_list, sample_sizes):
    df.to_csv("mean_changes/size_"+str(sample_size), header=True, index=False)
```

## Standard deviation difference detection

### Simulation

```python
sample_sizes = [15, 30, 50, 100, 150, 200, 500, 1000]

min_scale = 0.01
max_scale = 3.01

granularity = 300
num_iterations = 1000

scales = np.linspace(min_scale, max_scale, granularity)

df_list = list()

for sample_size in sample_sizes:
    df = pd.DataFrame(data=scales, columns=["std_data"])
    
    for i in range(num_iterations):
        ks_results = list()
        
        for scale in scales:
            s0 = np.random.normal(loc=0, scale=1, size=sample_size)
            s_test = np.random.normal(loc=0, scale=scale, size=sample_size)
            ks_results.append(stats.ks_2samp(s0, s_test)[1])
            
        df["iter_"+str(i)] = ks_results
        
    df["mean_res"] = df[df.columns[1:]].mean(axis=1)
    df["std_res"] = df[df.columns[1:]].std(axis=1)
    df_list.append(df)
```

### Plotting the results

```python
for df in df_list:
    fig, ax = plt.subplots(figsize=(25, 5))

    plt.errorbar(df["std_data"], 
                 df["mean_res"], 
                 yerr=df["std_res"], 
                 elinewidth=0.7,
                 linestyle="--", 
                 marker=".", ecolor="red", alpha=0.5, color="black")
    
    ax.fill_between([min_scale, max_scale], [0.01, 0.01], [0, 0], alpha=0.2, label="0.01")
    ax.fill_between([min_scale, max_scale], [0.05, 0.05], [0.01, 0.01], alpha=0.2, label="0.05")
    ax.fill_between([min_scale, max_scale], [0.1, 0.1], [0.05, 0.05], alpha=0.2, label="0.1")
    
    ax.set_xlim((min_scale, max_scale))
    ax.set_ylim((0, 1))
    
    ax.set_xlabel("Normal distribution standard deviation")
    ax.set_ylabel("p-value")
    ax.legend(title="Confidence levels")
    
plt.show()
```

### Saving the data to files

```python
for df, sample_size in zip(df_list, sample_sizes):
    df.to_csv("std_changes/size_"+str(sample_size), header=True, index=False)
```
