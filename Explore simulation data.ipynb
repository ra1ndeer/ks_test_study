# Explore simulation data

## Necessary imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path
from collections import OrderedDict
from mpl_toolkits.mplot3d import Axes3D

sns.set_style("whitegrid")
```

## Exploring the effect of changes in the mean of normally distributed random samples

```python
mean_simulation_path = Path("simulation_data/mean_changes")
simulation_files = [p.name for p in mean_simulation_path.glob("*")]

df_mean_dict = dict()

for simulation in simulation_files:
    df = pd.read_csv(str(mean_simulation_path / simulation))
    df_mean_dict[int(simulation.split("_")[1])] = df
    
df_mean_dict = OrderedDict(sorted(df_mean_dict.items()))
```

```python
for name, df in df_mean_dict.items():
    fig, ax = plt.subplots(figsize=(25, 5))

    plt.errorbar(df["mean_data"], 
                 df["mean_res"], 
                 yerr=df["std_res"], 
                 elinewidth=0.7,
                 linestyle="--", 
                 marker=".", ecolor="red", alpha=0.5, color="black")
    
    
    ax.fill_between([df["mean_data"].min(), df["mean_data"].max()], [0.1, 0.1], [0.05, 0.05], alpha=0.2, label="0.1")
    ax.fill_between([df["mean_data"].min(), df["mean_data"].max()], [0.05, 0.05], [0.01, 0.01], alpha=0.2, label="0.05")
    ax.fill_between([df["mean_data"].min(), df["mean_data"].max()], [0.01, 0.01], [0, 0], alpha=0.2, label="0.01")
    
    ax.set_xlim((df["mean_data"].min(), df["mean_data"].max()))
    ax.set_ylim((0, 1))
    
    ax.set_xlabel("Normal distribution mean")
    ax.set_ylabel("p-value")
    ax.legend(title="Confidence levels")
    ax.set_title("KS test p-value vs. changes in mean for sample size {}".format(name))
    
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(25, 5))

for name, df in df_mean_dict.items():
    
    ax.plot(df["mean_data"],
            df["mean_res"],
            linestyle="--",
            linewidth=0.7,
            marker=".",
            alpha=0.5, label="sample size: {}".format(name))

ax.fill_between([df["mean_data"].min(), df["mean_data"].max()], [0.1, 0.1], [0.05, 0.05], alpha=0.2, label="confidence level: 0.1")
ax.fill_between([df["mean_data"].min(), df["mean_data"].max()], [0.05, 0.05], [0.01, 0.01], alpha=0.2, label="confidence level: 0.05")
ax.fill_between([df["mean_data"].min(), df["mean_data"].max()], [0.01, 0.01], [0, 0], alpha=0.2, label="confidence level: 0.01")
    
ax.set_xlim((df["mean_data"].min(), df["mean_data"].max()))
ax.set_ylim((0, 0.7))
    
ax.set_xlabel("Mean")
ax.set_ylabel("p-value")
ax.legend()
ax.set_title("KS test p-value vs. changes in mean by sample size")
    
plt.show()
```

```python
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111, projection='3d')

for ssize, df in df_mean_dict.items():
    ax.plot(df["mean_data"], 
            df["mean_res"], 
            ssize, 
            zdir="y", 
            marker=" ", 
            linestyle="-", 
            alpha=0.5, 
            linewidth=3, 
            label="Sample size: {}".format(str(ssize)))
    
ax.set_ylim((1000, 0))

ax.set_xlabel("Mean")
ax.set_ylabel("Sample size")
ax.set_zlabel("p-value")

ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False

ax.xaxis.pane.set_edgecolor('w')
ax.yaxis.pane.set_edgecolor('w')
ax.zaxis.pane.set_edgecolor('w')

ax.legend(loc=(0.2, 0.5))

plt.show()
```

## Exploring the effect of changes in the standard deviation of normally distributed random samples

```python
std_simulation_path = Path("simulation_data/std_changes")
simulation_files = [p.name for p in std_simulation_path.glob("*")]

df_std_dict = dict()

for simulation in simulation_files:
    df = pd.read_csv(str(std_simulation_path / simulation))
    df_std_dict[int(simulation.split("_")[1])] = df
    
df_std_dict = OrderedDict(sorted(df_std_dict.items()))
```

```python
for name, df in df_std_dict.items():
    fig, ax = plt.subplots(figsize=(25, 5))

    plt.errorbar(df["std_data"], 
                 df["mean_res"], 
                 yerr=df["std_res"], 
                 elinewidth=0.7,
                 linestyle="--", 
                 marker=".", ecolor="red", alpha=0.5, color="black")

    ax.fill_between([df["std_data"].min(), df["std_data"].max()], [0.1, 0.1], [0.05, 0.05], alpha=0.2, label="0.1")
    ax.fill_between([df["std_data"].min(), df["std_data"].max()], [0.05, 0.05], [0.01, 0.01], alpha=0.2, label="0.05")
    ax.fill_between([df["std_data"].min(), df["std_data"].max()], [0.01, 0.01], [0, 0], alpha=0.2, label="0.01")
    
    ax.set_xlim((df["std_data"].min(), df["std_data"].max()))
    ax.set_ylim((0, 1))
    
    ax.set_xlabel("Normal distribution standard deviation")
    ax.set_ylabel("p-value")
    ax.legend(title="Confidence levels")
    ax.set_title("KS test p-value vs. changes in standard deviation for sample size {}".format(name))
    
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(25, 5))

for name, df in df_std_dict.items():
    
    ax.plot(df["std_data"],
            df["mean_res"],
            linestyle="--",
            linewidth=0.7,
            marker=".",
            alpha=0.5, label="sample size: {}".format(name))
            #color="black")
    
ax.fill_between([df["std_data"].min(), df["std_data"].max()], [0.1, 0.1], [0.05, 0.05], alpha=0.2, label="confidence level: 0.1")
ax.fill_between([df["std_data"].min(), df["std_data"].max()], [0.05, 0.05], [0.01, 0.01], alpha=0.2, label="confidence level: 0.05")
ax.fill_between([df["std_data"].min(), df["std_data"].max()], [0.01, 0.01], [0, 0], alpha=0.2, label="confidence level: 0.01")

    
ax.set_xlim((df["std_data"].min(), df["std_data"].max()))
ax.set_ylim((0, 1))
    
ax.set_xlabel("Standard deviation")
ax.set_ylabel("p-value")
ax.legend()
ax.set_title("KS test p-value vs. changes in standard deviation by sample size")
    
plt.show()
```
